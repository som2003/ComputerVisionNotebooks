```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(img)
plt.show()
```

# cv2.warpPerspective() - 3x3 transformation matrix
# cv2.warpAffine() - 2x3 transformation matrix

```python
#Translation
#transformation matrix
M=np.float32([[1,0,50],
              [0,1,50],
              [0,0,1]])
rows,cols,dim=img.shape
translated_img=cv2.warpPerspective(img,M,(cols,rows))
plt.axis('off')
plt.imshow(translated_img)
plt.show()
#plt.imsave("viratkohli_translated.jpg",translated_img)
```

```python
plt.imsave("D:\\Downloads\\viratkohli_translated.jpg",translated_img)
```

```python
#Scaling
#transformation matrix
M1=np.float32([[2,0,0],
               [0,3,0],
               [0,0,1]])
scaled_img=cv2.warpPerspective(img,M1,(cols*2,rows*2))
plt.axis('off')
plt.imshow(scaled_img)
plt.show()
plt.imsave("D:\\Downloads\\viratkohli_scaled.jpg",scaled_img)
```

```python
#Shearing
#transformation matrix - shear in x axis
Mx=np.float32([[1,0.5,0],
               [0,1,0],
               [0,0,1]])
#transformation matrix - shear in y axis
My=np.float32([[1,0,0],
               [0.5,1,0],
               [0,0,1]])
shearedX_img=cv2.warpPerspective(img,Mx,(int(cols*1.5),int(rows*1.5)))
shearedY_img=cv2.warpPerspective(img,My,(int(cols*1.5),int(rows*1.5)))
plt.axis('off')
plt.imshow(shearedX_img)
plt.imshow(shearedY_img)
plt.show()
plt.imsave("D:\\Downloads\\viratkohli_shearedX.jpg",shearedX_img)
plt.imsave("D:\\Downloads\\viratkohli_shearedY.jpg",shearedY_img)
```

```python
#Reflection
#transformation matrix - x axis reflection
M2=np.float32([[1,0,0],
               [0,-1,rows],
               [0,0,1]])
#transformation matrix - y axis reflection
M3=np.float32([[-1,0,cols],
               [0,1,0],
               [0,0,1]])
reflectedX_img=cv2.warpPerspective(img,M2,(int(cols),int(rows)))
plt.axis('off')
plt.imshow(reflectedX_img)
plt.show()
plt.imsave("D:\\Downloads\\viratkohli_reflectedX.jpg",reflectedX_img)
```

```python
reflectedY_img=cv2.warpPerspective(img,M3,(int(cols),int(rows)))
plt.axis('off')
plt.imshow(reflectedY_img)
plt.show()
plt.imsave("D:\\Downloads\\viratkohli_reflectedY.jpg",reflectedY_img)
```

```python
#Rotation
angle=np.radians(10)
M4=np.float32([[np.cos(angle),-np.sin(angle),0],
              [np.sin(angle),np.cos(angle),0],
              [0,0,1]])
rotated_img=cv2.warpPerspective(img,M4,(int(cols),int(rows)))
plt.axis('off')
plt.imshow(rotated_img)
plt.show()
plt.imsave("D:\\Downloads\\viratkohli_rotated.jpg",rotated_img)
```

```python
#cropping
#within the range of rows and cols
cropped_img=img[200:600,300:800]
plt.axis('off')
plt.imshow(cropped_img)
plt.show()
plt.imsave("D:\\Downloads\\viratkohli_cropped.jpg",cropped_img)
```

```python
#Affine transformation - parallel lines remain parallel
pts1=np.float32([[50,50],[200,50],[50,200]]) #(source,destination)points with which affine matrix is calculated
pts2=np.float32([[10,100],[200,50],[100,250]])
M5=cv2.getAffineTransform(pts1,pts2)
affine_img=cv2.warpAffine(img,M5,(cols,rows))
plt.axis('off')
plt.imshow(affine_img)
plt.show()
plt.imsave("D:\\Downloads\\viratkohli_affine.jpg",affine_img)
```

```python
#projective transformation - perspective transformation - homography
#arbitrary transformation matrix
pts1=np.float32([[50,60],[370,46],[30,387],[456,460]])
pts2=np.float32([[0,0],[250,0],[0,250],[250,250]])
M6=cv2.getPerspectiveTransform(pts1,pts2)
perspective_img=cv2.warpPerspective(img,M6,(400,400))
plt.axis('off')
plt.imshow(perspective_img)
plt.show()
plt.imsave("D:\\Downloads\\viratkohli_perspective.jpg",perspective_img)
```
